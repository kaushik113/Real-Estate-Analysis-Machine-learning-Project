```python
import numpy as np
from sklearn.model_selection import train_test_split

# Custom split function
def split_train_test(data, test_ratio):
    np.random.seed(42)
    shuffle = np.random.permutation(len(data))
    print(shuffle)
    test_set_size = int(len(data) * test_ratio)
    test_indices = shuffle[:test_set_size]
    train_indices = shuffle[test_set_size:]
    return data.iloc[train_indices], data.iloc[test_indices]

# Using custom split function
train_set_custom, test_set_custom = split_train_test(house, test_ratio=0.2)
print(f"Custom Split - \nRows in train set: {len(train_set_custom)}\nRows in test set: {len(test_set_custom)}\n")
# Using sklearn's train_test_split function

train_set, test_set = train_test_split(house, test_size=0.2, random_state=42)
print(f"Sklearn Split - \nRows in train set: {len(train_set)}\nRows in test set: {len(test_set)}\n")
```

### Example in case of data types like "CHAS" where majority are 0 and some are 1, if the data has not seen "CHAS : 1" , then it will have false preassumptions.SO to prevent to do STRATIFIED LEARNING.


```python
from sklearn.model_selection import StratifiedShuffleSplit

# Create the stratified shuffle split object
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)

# Perform the split
for train_index, test_index in split.split(house, house['CHAS']):
    strat_train_set = house.loc[train_index]
    strat_test_set = house.loc[test_index]
```

```python
# Display information about the stratified test set
strat_test_set.info()

# Display value counts of the 'CHAS' column in the test set
print("CHAS value counts in stratified test set:")
print(strat_test_set['CHAS'].value_counts())

# Display value counts of the 'CHAS' column in the training set
print("CHAS value counts in stratified training set:")
print(strat_train_set['CHAS'].value_counts())
```

# Scikit-learn Design

Divided into three objevts:
1.Estimator - fits data set and transform data values. example - imputer
2.Transformers - it takes leaning from fit(). Ex : fit_transform()
3.Predictor - LinearRegression, 2 comman function : fit() & predict().

# Creating a Pipeline

```python
from sklearn.pipeline import Pipeline 
```

## Feature Scaling
1.Min-max scalling (normalization)
 (value - min) / (max -min) 
 Sklearn provides a class called MinMacScaler
 
2. Standardization
  (value - mean) / std
  Sklearn provides a class called [Standard Scaler] -> variance is set to 1

```python
from sklearn.pipeline import Pipeline 
from sklearn.preprocessing import StandardScaler
my_pipeline = Pipeline([
    ('imputer' , SimpleImputer(strategy = "median")),
    #..
    #..
    ('std_scaler' , StandardScaler() ),
])
#this applies fit transform in this
#now this apply fit tranform to this 
```

```python
housing_num_tr = my_pipeline.fit_transform(house)
```

```python
housing_num_tr
housing_num_tr.shape
```

# Selecting a desired model for real estate

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
model = LinearRegression()
# model= DecisionTreeRegressor()
model = RandomForestRegressor()
model.fit(housing_num_tr,housing_labels)
```

```python
some_data = house.iloc[ :5]
```

```python
some_labels = housing_labels.iloc[:5]
```

```python
prepared_data = my_pipeline.transform(some_data)
```

```python
model.predict(prepared_data)
```

```python
from sklearn.metrics import mean_squared_error
housing_predictions = model.predict(housing_num_tr)
mse = mean_squared_error(housing_labels, housing_predictions)
rmse = np.sqrt(mse)
```

```python
rmse
```

# Evaluation technique - Cross Validation

```python
from sklearn.model_selection import cross_val_score
scores =cross_val_score(model  , housing_num_tr, housing_labels, scoring = "neg_mean_squared_error", cv=10)
rmse_scores = np.sqrt(-scores)
```

```python
rmse_scores
```

```python
def print_scores(scores):
    print("Scores : ", scores)
    print("Mean :", scores.mean())
    print("Standard deviation :",scores.std())
```

```python
print_scores(rmse_scores)
```
